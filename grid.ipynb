3.Explain the use of grid-auto-row and grid-auto-column using code examples?
Ans:

grid-auto-rows and grid-auto-columns in CSS Grid
These two properties are part of the implicit grid generation in CSS Grid. When you place grid items, if you don't explicitly define enough tracks (rows or columns) with grid-template-rows or grid-template-columns to accommodate all items, the grid will implicitly create new tracks. grid-auto-rows and grid-auto-columns allow you to control the size of these implicitly created tracks.
1. grid-auto-rows
Purpose: This property specifies the size of implicitly created grid rows.
When it's used:
 * When grid items are placed into rows that have not been explicitly defined by grid-template-rows. This often happens when you use grid-auto-flow: column; or explicitly place items into rows beyond your grid-template-rows definition.

Syntax:
grid-auto-rows: track-size>+;
Where track-size> can be a length (e.g., 100px), a percentage, fr unit (e.g., 1fr), min-content, max-content, auto, or a minmax() function.
Example Scenario: Imagine a gallery where you define fixed columns, but you want new rows that are created automatically to have a specific minimum height.
Code Example for grid-auto-rows:

HTML (index.html):


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>grid-auto-rows Example</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <h1>grid-auto-rows Example</h1>
    <p>We define 3 explicit columns and set grid-auto-rows to 120px. The grid will automatically add rows as needed, and those new rows will be 120px tall.</p>
    <div class="grid-container auto-rows-example">
        <div class="grid-item">1</div>
        <div class="grid-item">2</div>
        <div class="grid-item">3</div>
        <div class="grid-item">4 (New Row)</div>
        <div class="grid-item">5 (New Row)</div>
        <div class="grid-item">6 (New Row)</div>
        <div class="grid-item">7 (Another New Row)</div>
        </div>
</body>
</html>
```

Explanation for grid-auto-rows example:
 * We define grid-template-columns: repeat(3, 1fr);, meaning we have 3 explicit columns.
 * We do not define grid-template-rows.
 * We add more than 3 grid items (e.g., 7 items).
 * The first 3 items will fill the first implicit row.
 * Item 4 will automatically create a new, implicit row. Because grid-auto-rows: 120px; is set, this new row (and any subsequent implicit rows) will have a height of 120px. Without grid-auto-rows, these implicit rows would default to auto, shrinking to fit their content.


2. grid-auto-columns
Purpose:

 This property specifies the size of implicitly created grid columns.
When it's used:
 * When grid items are placed into columns that have not been explicitly defined by grid-template-columns. This usually happens when grid-auto-flow: column; is set, or when you explicitly place items into columns beyond your grid-template-columns definition.

Syntax:
grid-auto-columns: track-size>+;
The track-size> values are the same as for grid-auto-rows.
Example Scenario: Imagine a horizontal layout where you define fixed rows, but you want new columns that are created automatically to have a specific width.
Code Example for grid-auto-columns:

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>grid-auto-columns Example</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <h1>grid-auto-columns Example</h1>
    <p>We define 2 explicit rows and set grid-auto-columns to 100px. We also set grid-auto-flow: column; so items fill columns first. The grid will automatically add columns as needed, and those new columns will be 100px wide.</p>
    <div class="grid-container auto-columns-example">
        <div class="grid-item">1</div>
        <div class="grid-item">2</div>
        <div class="grid-item">3 (New Column)</div>
        <div class="grid-item">4 (New Column)</div>
        <div class="grid-item">5 (Another New Column)</div>
        <div class="grid-item">6 (Another New Column)</div>
    </div>
</body>
</html>

CSS (style.css - add to existing file or replace for testing this example):
body {
    font-family: Arial, sans-serif;
    margin: 20px;
    background-color: #f4f4f4;
}

h1 {
    text-align: center;
    color: #333;
}

p {
    text-align: center;
    max-width: 600px;
    margin: 10px auto 30px;
    line-height: 1.5;
    color: #555;
}

.grid-container {
    display: grid;
    gap: 10px;
    border: 1px solid #ccc;
    padding: 15px;
    background-color: #fff;
    max-width: 800px;
    margin: 0 auto;
}

.grid-item {
    display: flex;
    justify-content: center;
    align-items: center;
    background-color: #aed6f1; /* Light blue */
    border: 1px solid #3498db; /* Darker blue */
    color: #21618c;
    font-size: 1.2em;
    font-weight: bold;
    padding: 10px;
    min-width: 80px; /* Ensure a minimum width for items */
}

/* Specific example for grid-auto-columns */
.auto-columns-example {
    /* Explicitly define 2 rows */
    grid-template-rows: repeat(2, 100px);

    /* Items will flow into columns first instead of rows */
    grid-auto-flow: column;

    /* Implicitly created columns will be 100px wide */
    grid-auto-columns: 100px;
}
```

 * We do not define grid-template-columns.
 * We set grid-auto-flow: column;. This is crucial because it changes the default behavior of grid item placement. Instead of filling rows left-to-right, top-to-bottom, items will fill columns top-to-bottom, left-to-right.
 * We add more than 2 grid items (e.g., 6 items).
 * The first 2 items will fill the first implicit column.
 * Item 3 will automatically create a new, implicit column. Because grid-auto-columns: 100px; is set, this new column (and any subsequent implicit columns) will have a width of 100px. Without grid-auto-columns, these implicit columns would default to auto, shrinking to fit their content.

5.Explain the difference between justify-items and justify-self using code examples?

ANS:Understanding justify-items and justify-self in CSS Grid
Both justify-items and justify-self are used for aligning grid items within their respective grid cells along the inline (row) axis. The key difference lies in where these properties are applied and what they affect:
 * justify-items: Applied to the Grid Container. It aligns all grid items inside that container along the inline axis.
 * justify-self: Applied to an individual Grid Item. It aligns that specific item within its own cell along the inline axis, overriding any justify-items value set on the container.
Think of it like this: justify-items is a "bulk" setting for all children, while justify-self is a "per-child" override.
1. justify-items
Purpose :
Aligns the content of grid items along the inline (row) axis within their grid cells, for all direct children of the grid container.
Applied to: The Grid Container.
Values:
 * start: Aligns items to the start of their cell's inline axis.
 * end: Aligns items to the end of their cell's inline axis.
 * center: Aligns items to the center of their cell's inline axis.
 * stretch (default): Stretches items to fill the entire width of their cell (if the item's intrinsic size allows).
 * left, right, space-around, space-between, space-evenly: (Less commonly used with justify-items directly for individual item alignment, more for justify-content).
Code Example for justify-items:
HTML (index.html):


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>justify-items Example</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <h1>justify-items Example</h1>
    <p>This property is applied to the <strong>grid container</strong> and affects the alignment of <strong>all grid items</strong> along the row (inline) axis within their cells.</p>

    <h2>justify-items: start;</h2>
    <div class="grid-container justify-items-start">
        <div class="grid-item">Item 1</div>
        <div class="grid-item">Item 2 long content</div>
        <div class="grid-item">Item 3</div>
    </div>

    <h2>justify-items: center;</h2>
    <div class="grid-container justify-items-center">
        <div class="grid-item">Item 1</div>
        <div class="grid-item">Item 2 long content</div>
        <div class="grid-item">Item 3</div>
    </div>

    <h2>justify-items: end;</h2>
    <div class="grid-container justify-items-end">
        <div class="grid-item">Item 1</div>
        <div class="grid-item">Item 2 long content</div>
        <div class="grid-item">Item 3</div>
    </div>

    <h2>justify-items: stretch; (Default behavior)</h2>
    <div class="grid-container justify-items-stretch">
        <div class="grid-item">Item 1</div>
        <div class="grid-item">Item 2 long content</div>
        <div class="grid-item">Item 3</div>
    </div>
</body>
</html>

CSS (style.css):
body {
    font-family: Arial, sans-serif;
    margin: 20px;
    background-color: #f4f4f4;
    text-align: center;
}

h1 {
    color: #333;
}

h2 {
    margin-top: 40px;
    color: #444;
}

p {
    max-width: 700px;
    margin: 10px auto 30px;
    line-height: 1.5;
    color: #555;
}

.grid-container {
    display: grid;
    grid-template-columns: repeat(3, 150px); /* Fixed width columns to show alignment */
    gap: 10px;
    border: 2px solid #3498db;
    padding: 15px;
    background-color: #e8f0fe;
    max-width: 600px;
    margin: 0 auto 30px;
    box-sizing: border-box; /* Include padding and border in element's total width and height */
}

.grid-item {
    background-color: #aed6f1;
    border: 1px solid #2874a6;
    color: #21618c;
    font-size: 1.1em;
    font-weight: bold;
    padding: 10px;
    /* Important: Remove default width/height to see stretch clearly */
    width: auto;
    height: auto;
}

/* Specific justify-items examples */
.justify-items-start {
    justify-items: start;
}

.justify-items-center {
    justify-items: center;
}

.justify-items-end {
    justify-items: end;
}

.justify-items-stretch {
    justify-items: stretch; /* Default behavior, but explicitly shown */
}
```

Explanation for justify-items example:
We've set grid-template-columns to fixed 150px widths. This ensures there's extra space within each cell for the grid items to be aligned.
 * justify-items: start;: All items will align to the left (start) of their 150px cell.
 * justify-items: center;: All items will center horizontally within their 150px cell.
 * justify-items: end;: All items will align to the right (end) of their 150px cell.
 * justify-items: stretch;: (Default) Items will expand to fill the entire 150px width of their cell, unless their content intrinsically prevents it (e.g., an image with max-width: none;). Notice how "Item 2 long content" stretches.


2. justify-self
Purpose:
 Aligns a specific grid item's content along the inline (row) axis within its own grid cell, overriding any justify-items value set on the container.
Applied to: An individual Grid Item.
Values:
 * start, end, center, stretch (default), left, right, auto.
Code Example for justify-self:

HTML (index.html - you can add this below the previous examples or replace them):


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>justify-self Example</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <h1>justify-self Example</h1>
    <p>This property is applied to an <strong>individual grid item</strong> and overrides the container's justify-items setting for that specific item.</p>

    <h2>Using justify-self to override justify-items: center;</h2>
    <div class="grid-container justify-items-center-override">
        <div class="grid-item">Default Center</div>
        <div class="grid-item justify-self-start">Start</div>
        <div class="grid-item justify-self-end">End</div>
        <div class="grid-item justify-self-stretch">Stretch</div>
        <div class="grid-item">Default Center</div>
    </div>
</body>
</html>

CSS (style.css - add to existing file):
/* ... (previous CSS for body, h1, h2, p, .grid-container, .grid-item) ... */

/* Specific example for justify-self */
.justify-items-center-override {
    justify-items: center; /* All items would normally be centered */
    grid-template-columns: repeat(5, 150px); /* 5 columns for more items */
}

/* Individual item overrides using justify-self */
.justify-self-start {
    justify-self: start; /* This item aligns to the start */
}

.justify-self-end {
    justify-self: end; /* This item aligns to the end */
}

.justify-self-stretch {
    justify-self: stretch; /* This item stretches */
}
```

Explanation for justify-self example:
 * We've applied justify-items: center; to the .grid-container. This means by default, all items within this container should be horizontally centered in their cells.
 * However, for grid-items with justify-self-start, justify-self-end, and justify-self-stretch classes, we apply justify-self to them.
 * You will see that these specific items ignore the justify-items: center; setting from their parent and instead align according to their own justify-self value. The "Default Center" items will still be centered because they don't have an overriding justify-self property.
Analogy:
Think of it like setting the default text alignment for a document versus aligning individual paragraphs:
 * justify-items is like setting text-align: center; on the <body> element. All text inside will be centered by default.
 * justify-self is like setting text-align: right; on a specific <p> tag. That paragraph will be right-aligned, even if the body wants everything centered.
In summary, justify-items is a property for the grid container to align all its items as a group along the inline axis, while justify-self is for individual grid items to align themselves within their own cell, potentially overriding the container's setting.
